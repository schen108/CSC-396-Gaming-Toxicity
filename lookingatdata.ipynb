```python
path = "/home/jovyan/.cache/kagglehub/datasets/simshengxue/league-of-legends-tribunal-chatlogs/versions/3/chatlogs.csv"
```

```python
dataset = pd.read_csv(path)
```

```python
dataset.head()
```

```python
df_dropped = dataset[['message','chatlog_id','champion_name']]
```

```python
df_dropped['message'] = df_dropped['message'].fillna('').astype(str)
```

```python
df_combined = df_dropped.groupby(['chatlog_id','champion_name'])['message'].apply(','.join).reset_index()
```

```python
print(df_combined)
```
